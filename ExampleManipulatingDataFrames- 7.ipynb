```python
import pandas as pd
```

```python
df = pd.read_csv("president.csv", index_col=0)
df.head()
```

```python
# let's seperate first name of president from its full name 
df["first"] = df["president"]
df["first"] = df["first"].replace("[ ].*", "", regex= True)
df.head()
```

```python
# We can do the same thing in short and faster way so let's erase the column "first" permanently
del(df["first"])
```

```python
df.head()
```

```python
def splitname(row):
    # Let's do for first name
    row["first"] = row["president"].split(" ")[0]
    # Let's do the same for last name
    row["last"] = row["president"].split(" ")[-1]
    return row;

df = df.apply(splitname, axis=1) # Here we can also do axis = "coloumn"
df.head()
```

```python
# Here we have more fast and easy way to do above thing 
del(df["first"])
del(df["last"])
```

```python
# Using regex we can do it fast
pattern= "(^[\w]*)(?:.* )([\w]*$)"
df["president"].str.extract(pattern).head()
```

```python
# Let's give names to the columns 
pattern= "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

names= df["president"].str.extract(pattern).head()
names
```

```python
# Now let's insert into main DataFrame
df["First"] = names["First"]
df["Last"] = names["Last"]
df.head(10)
```

```python
# Let's go to start column
df["start"] = df["start"].str.extract("([\w]*} [\w]{1,2}, [\w]{4})")
df["start"].head()
```

```python
# Pandas also has great library date/time let's use it 
df["start"] = pd.to_datetime(df["start"])
df["start"].head()
```
