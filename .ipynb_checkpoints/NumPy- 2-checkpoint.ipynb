# Array Operations

```python
# In array we can do many things such as addtion, subtraction, multiplication, use boolean values which are 
# binary openrations. And many other things like matrix manipulation such as product, transverse, inverse, etc

#arithematic operators in array apply element wise

m = np.array([10,2,30])
n = np.array([1,2,3])

p = m*n;
add = m+n;
print(add)
print(p)
```

```python
#I moved down to the United States about six years ago from Canada. In Canada, 
#we use Celsius for temperatures and my wife still hasn't converted to the US system, which uses Fahrenheit.
#With numpy I could easily convert a number of Fahrenheit C = ((F-32)*(5/9))

far = np.array([0,-10,-15,1])

cel = (far-32)*(5/9)
print(cel)
```

```python
#boolean array
cel > -20
```

```python
A = np.array([[1,1],[0,1]])
B = np.array([[2,0],[3,4]])

art = A*B #This is arithematic product which multiply element wise
pro = (A@B) #This is actual matrix product which can be done using "@" or "np.dot(A,B)" opearator

print(art)
print(pro)
```

```python
# NumPy has some intresting aggregation functions such as sum(), max(), min(), mean(), etc
print(pro.sum())
print(pro.min())
print(pro.max())
print(pro.mean())
```

```python
# Easy way to creat a array
g = np.arange(1,16,1).reshape(3,5)
print(g)
```

# NumPy for Images

```python
from PIL import Image
from IPython.display import display
```

```python
im = Image.open("imgru.jpg")
display(im)
```

```python
array2 = np.array(im)
print(array2.shape)
array2
```

```python
mask = np.full(array2.shape, 255)
mask
```

```python
# For black and white images, black is stored as 0 and white is stored as 255. So, to invert the image we can use 
# the numpy array
modified_array = array2-mask

#lets convert all negative values to positive
modified_array = modified_array*-1

modified_array = modified_array.astype(np.uint8)
modified_array 
```

```python
display(Image.fromarray(modified_array)) # We use fromarray() function to convert numpy array into an object 
                                         # jupyter can render
```

# Indexing, Slicing and Iterating

```python
# For one dimentional array
s = np.array([1,2,3])
s[2]
```

```python
# For multidimentional array
m = np.array([[1,2],[3,4],[5,6]])
print(m.shape)
m
```

```python
m[1,1] # Here first is row and second is coloumn and we start with 0 indexing
```

```python
# If we want to creat list of elements 1,4,6 from m array then
np.array([m[0,0], m[1,1], m[2,1]])
```

```python
# Another way of indexing is using the zip of slices of position of m
print(m[[0,1,2],[0,1,1]])
```

# Boolean Indexing

```python
# From the above array m 
m>5
```

```python
# Here we can creat the one dimentional array of the elelments which has true vlaues in array m
print(m[m>5])
```

# Slicing

```python
# In one dimention 
o = np.array([1,2,3,4,5,6])
o[:2] # Here 2 is exclusive
```

```python
# for multidimentional array 
mu = np.array([[11,12,13,14],[21,22,23,24],[31,32,33,34]])
print(mu)
```

```python
mu[:2, 1:3] # Here in multidimentional array the first argument is for row and second is for coloumn
```

```python
sub_array = mu[:2, 1:3]
print("The [0,0] of sub_array is: ", sub_array[0,0]);
sub_array[0,0] = sub_array[1,1]
print("After changing the value is: ", sub_array[0,0])
print(sub_array)
```
