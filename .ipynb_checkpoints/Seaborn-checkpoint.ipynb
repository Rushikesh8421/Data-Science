# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips) #Here the y axes is fixed on we have to provide x
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)  #hue parameter is also used here 
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# Combining two graphs
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
# the .catplot is common plot in which we have use kind and assing the type of plot we want
```

# Matrix Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

## .heatmap()

```python
tc = tips.corr() #we have to make it in matrix form so .corr() give both row and col same
tc               # same name and we can get access by using cols and rows particularly
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')#annot give values on box,cmap give colorfullness
```

```python
flights
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='black',linewidths='3')
```

```python
sns.heatmap(fp)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

## Grid

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.pairplot(iris)
g
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.histplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

# Regression plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
scatter_kws={'s':100}) #scatter-kws decides the size of graph plots
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=8)
```
