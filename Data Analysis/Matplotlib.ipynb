```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# %matplotlib inline sets the backend of matplotlib to the 'inline' backend: 
# With this backend, the output of plotting commands is displayed inline within frontends 
# like the Jupyter notebook, directly below the code cell that produced it. 
# The resulting plots will then also be stored in the notebook document.
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

# Functional

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.show() this is used to show the output if we are using IDE other than Jupyter notebook
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
# Here in .add_axes([from left,from bottom, width, lenght])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Large Plot")

axes2.plot(y,x)
axes2.set_title("Small Plot")
```

# Subplots using object orentation

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First graph')

axes[1].plot(y,x)
axes[1].set_title('Second graph')
```

# Figure size and DPI

```python
fig = plt.figure(figsize=(8,2)) #i.e 3 inches by 2 inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4))

axes[0].plot(x,y)

axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('My_pic',dpi=200) #dot per inches....It saves the graph image in pc
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='Y Cubed')

ax.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='red',lw=3,alpha=1,ls='--',marker='o',markersize=30,
          markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
#lw is linewidth, alpha give the transparency, linestyle
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='blue',lw='2',ls='--')
axes.set_xlim([0,1]) # It takes the slice of x axis
axes.set_ylim([0,2]) #It takes the slice of y axis
```
