```python
df = pd.read_csv("in.csv")
df.head(10)
```

```python
df = df.set_index("admin_name")
df = df.sort_index()
df.head(20)
```

```python
# as we see above all the values of index are not unique so we set 2 index
df = df.reset_index()
df = df.set_index(["admin_name","city"])
df.head(10)
```

```python
df = df.fillna(method="ffill") # This funciton fills the values which are not in dataset,same value for particular 
# column, index, etc are filled in place of NaN
df.head()
```

```python
df = pd.DataFrame({"A": [1,1,2,3,4,5],
                 "B": [3,4,3,8,9,10],
                  "C": ['a','b','c','d','e','f']})
df
```

```python
# .replace(what we want to put, what we want to replace with)
df.replace(1,100)
```

```python
# .replace() for multiple values replacement 
df.replace([1,3],[100,300])
```

```python
# we can also replace string using regex
df = pd.read_csv("data.csv")
df.head()
```

```python
# just insert pattern to replace in to_replace parameter
df.replace(to_replace="Current prices$", value="Present cost", regex=True)
```
